```python
def setStartDate(year,month,date) :
    yearElement = Select(driver.find_element_by_xpath("//*[@id='START_DATEY']"))
    monthElement = Select(driver.find_element_by_xpath("//*[@id='START_DATEM']"))
    dateElement = Select(driver.find_element_by_xpath("//*[@id='START_DATED']"))
    
    monthElement.select_by_visible_text("01")
    dateElement.select_by_visible_text("01")
   

    yearElement.select_by_visible_text(year)
    monthElement.select_by_visible_text(month)
    dateElement.select_by_visible_text(date)
```

```python
def setEndDate(year,month,date) :
    yearElement = Select(driver.find_element_by_xpath("//*[@id='END_DATEY']"))
    monthElement = Select(driver.find_element_by_xpath("//*[@id='END_DATEM']"))
    dateElement = Select(driver.find_element_by_xpath("//*[@id='END_DATED']"))
    
    monthElement.select_by_visible_text("01")
    dateElement.select_by_visible_text("01")
    
    yearElement.select_by_visible_text(year)
    monthElement.select_by_visible_text(month)
    dateElement.select_by_visible_text(date)
```

```python
def findStartAndEndDate(lastEndDate) :
    
    dateNum = [31,28,31,30,31,30,31,31,30,31,30,31]
    
    startYear  = int(lastEndDate[0])
    startMonth = int(lastEndDate[1])
    startDate  = int(lastEndDate[2]) + 1
    if startDate > dateNum[startMonth - 1] :
        startDate = startDate -  dateNum[startMonth - 1]
        startMonth += 1
        if startMonth > 12 :
            startMonth = 1
            startYear += 1  
    endYear = startYear
    endMonth = startMonth
    endDate = startDate + 6
    if endDate > dateNum[endMonth - 1] :
        endDate = endDate - dateNum[endMonth - 1]
        endMonth += 1
        if endMonth > 12 :
            endMonth = 1
            endYear += 1
            
    strStartYear  = str(startYear)
    strStartMonth = str(startMonth)
    strStartDate  = str(startDate)
    if int(startMonth / 10) <1 :
        strStartMonth = "0" + strStartMonth
    if int(startDate / 10) <1 :
        strStartDate = "0" +strStartDate
    strStarDateList=  [strStartYear,strStartMonth,strStartDate] 
    
    strEndYear  = str(endYear)
    strEndMonth = str(endMonth)
    strEndDate  = str(endDate)
    if endMonth / 10 <1 :
        strEndMonth = "0" + strEndMonth
    if endDate / 10 <1 :
        strEndDate = "0" +strEndDate
    strEndDateList=  [strEndYear,strEndMonth,strEndDate]
    return strStarDateList, strEndDateList
```

```python
def date_to_week() :
    initYear = "2016"
    initMonth = "12" 
    initDate = "31"
    lastEndDate = [initYear,initMonth,initDate]
    date_to_week_list = []
    cnt = 0
    while True :
        s,e = findStartAndEndDate(lastEndDate) 
        date_to_week_list.append([s,e])
        lastEndDate = e
        cnt += 1
        if cnt == 150 :
            break
    return date_to_week_list
```

```python
def readExchangeFile(dirPath) :
    fileList = listdir(dirPath)
    countrySize = 43
    fileSize = len(fileList) +1
    exchangeRateMatrix = [[0]*countrySize]*fileSize
    cnt = 0
    for filename in fileList :
        wb = xlrd.open_workbook(dirPath + filename)
        ws=wb.sheet_by_index(0)
        
        #  extract date from file
        strDate = ws.col_values(0)[0]
        strStarDateList ,strEndDateList = seperateStringToDate(strDate)
        date_to_weeks = date_to_week()
        # find week index from date
        idx = findIndexWithDate(strStarDateList,date_to_weeks)
        exchangeRate = ws.col_values(7)[3:]
        exchangeLen = len(exchangeRate)
        if exchangeLen == 47 :
            exchangeRate = exchangeRate[:-4]
        elif exchangeLen == 45 :
            exchangeRate = exchangeRate[:-2]
        elif exchangeLen == 0 :
            exchangeRate = exchangeRateMatrix[idx-1]
        exchangeRateMatrix[idx] = exchangeRate
        

    print(len(exchangeRateMatrix))
    beforeRate = exchangeRateMatrix[0]
    for idx in range(len(exchangeRateMatrix)) :
        if idx + 1 < fileSize :
            afterRate =  exchangeRateMatrix[idx + 1]
            tmp = [0]*countrySize
            for idx2 in range(countrySize) :
                tmp[idx2] = float(afterRate[idx2]) / float(beforeRate[idx2]) - 1.0
            exchangeRateMatrix[idx] = tmp
    return exchangeRateMatrix
            
```

```python
import win32com.client as win32
dname = 'D:/GoogleDiver/4학년1학기/nlp/termproject/source/exchangeRate/'
fileList = listdir(dname)
for filename in fileList :
    fname = filename
    excel = win32.gencache.EnsureDispatch('Excel.Application')
    wb = excel.Workbooks.Open(dname + fname)

    wb.SaveAs(fname+"x", FileFormat = 51)    #FileFormat = 51 is for .xlsx extension
    wb.Close()                               #FileFormat = 56 is for .xls extension
excel.Application.Quit()
```

```python
def seperateStringToDate(strDate) :
    strDate = strDate.replace(" ","")
    strDate = strDate.replace("~",".")
    strDate = strDate[:-7]
    strList = strDate.split(".")
    strStarDateList=  [strList[0],strList[1],strList[2]]
    strEndDateList=   [strList[3],strList[4],strList[5]]
    return strStarDateList ,strEndDateList
```

```python
def findIndexWithDate(strStarDateList,date_to_week) :
    idx = 0
    for item in date_to_week :
        if item[0] == strStarDateList :
            return idx
        idx += 1
    return idx
```
